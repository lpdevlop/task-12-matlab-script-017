```python
original_image = 'Original.png';
watermark_image = 'Watermark.png';
key_image       = 'key.png';         
watermarked_out = 'Watermarked.png';  

original_image_gray = imread(original_image);
if size(original_image_gray,3) == 3
    original_image_gray = rgb2gray(original_image_gray);
end
original_image_gray = uint8(original_image_gray);
totalPixels = numel(original_image_gray);

watermark_image_gray = imread(watermark_image);
if size(watermark_image_gray,3) == 3-
    watermark_image_gray = rgb2gray(watermark_image_gray);
end
watermark_image_gray = imbinarize(watermark_image_gray);
bits = watermark_image_gray(:);
numBits = length(bits);

key_image_gray = imread(key_image);
if size(key_image_gray,3) == 3
    key_image_gray = rgb2gray(key_image_gray);
end

if ~islogical(key_image_gray)
    key_image_gray = imbinarize(key_image_gray); 
end

kbits = key_image_gray(:)';
seed = mod(sum(double(kbits) .* (1:length(kbits))), 2^31-1);
rng(seed);  

permPositions = randperm(totalPixels, numBits);


watermark_image = original_image_gray;   

for i = 1:numBits
    pos = permPositions(i);
    watermark_image(pos) = bitset(watermark_image(pos), 1, bits(i));   
end

imwrite(watermark_image, watermarked_out);
fprintf("Watermark embedded%s\n", watermarked_out);


watermarked_out_image = imread(watermarked_out);
if size(watermarked_out_image,3) == 3
    watermarked_out_image = rgb2gray(watermarked_out_image);
end

exBits = zeros(numBits,1);

for i = 1:numBits
    pos = permPositions(i);
    exBits(i) = bitget(watermarked_out_image(pos), 1);
end

recovered-image = reshape(exBits, size(watermarked_out_image));
imwrite(recovered-image, 'Recovered_Watermark.png');

figure; 
subplot(1,3,1); imshow(I); title('Original');
subplot(1,3,2); imshow(I2); title('Watermarked');
subplot(1,3,3); imshow(Recovered); title('Recovered Watermark');
```
